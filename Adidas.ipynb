```python
import matplotlib.pyplot as plt
```

```python
df.groupby(['region'])['total_sales'].sum()
```

```python
sales_by_region = df.groupby('region')['total_sales'].sum().reset_index()


sales_by_region['total_sales_sum_million'] = sales_by_region['total_sales'] / 1_000_000


plt.bar(sales_by_region['region'], sales_by_region['total_sales_sum_million'], width=0.5)


plt.xlabel('Region')
plt.ylabel('Total Sales in  (Milion)')
plt.title('Total Sales by Region')

plt.show()
```

```python
sales_method_details = (
    df.groupby('sales_method')['total_sales']
      .sum()
      .reset_index()
)


sales_method_details['total_sales_sum_million'] = sales_method_details['total_sales'] / 1_000_000


plt.bar(sales_method_details['sales_method'], sales_method_details['total_sales_sum_million'], width=0.5)

plt.xlabel('Sales Method')
plt.ylabel('Total Sales (Million)')
plt.title('Total Sales by Sales Method')


plt.tight_layout()
plt.show()
```

```python
product_details = (
    df.groupby('product')['total_sales']
      .sum()
      .reset_index()
)


product_details['total_sales_sum_million'] = product_details['total_sales'] / 1_000_000


plt.bar(product_details['product'], product_details['total_sales_sum_million'], width=0.5)

plt.xlabel('product')
plt.ylabel('Total Sales (Million)')
plt.title('Total Sales by product')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
A = (
    df.groupby('retailer')['total_sales']
      .sum()
      .reset_index()
)


A['total_sales_sum_million'] = A['total_sales'] / 1_000_000


plt.bar(A['retailer'], A['total_sales_sum_million'], width=0.5)

plt.xlabel('Retailer')
plt.ylabel('Total Sales (Million)')
plt.title('Total Sales by retailer')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
